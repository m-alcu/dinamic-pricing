```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable
```

```python
# Let's make some data for a linear regression.
A = 3.1415926
b = 2.7189351
error = 0.1
N = 100 # number of data points
```

```python
# Data
X = Variable(torch.randn(N, 1))
```

```python
# (noisy) Target values that we want to learn.
t = A * X + b + Variable(torch.randn(N, 1) * error)
```

```python
# Creating a model, making the optimizer, defining loss
loss_fn = nn.MSELoss()
```

```python
A_ = Variable(torch.randn(1, 1), requires_grad=True)
b_ = Variable(torch.randn(1), requires_grad=True)
optimizer = optim.SGD([A_, b_], lr=1e-1)
```

```python
# Run training
niter = 50
for _ in range(0, niter):
    optimizer.zero_grad()
    predictions = torch.matmul(X, A_) + b_
    loss = loss_fn(predictions, t)
    loss.backward()
    optimizer.step()

    print("-" * 50)
    print("error = {}".format(loss.data))
    print("learned A_ = {}".format(A_))
    print("learned b_ = {}".format(b_))
```
